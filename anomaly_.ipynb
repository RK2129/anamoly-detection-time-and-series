```python
# Cell 5: Question 4 - Stationarity
print("=" * 80)
print("QUESTION 4: Stationarity in Time Series")
print("=" * 80)

answer4 = """
STATIONARITY IN TIME SERIES:

DEFINITION:
A time series is stationary when its statistical properties (mean, variance, covariance)
remain constant over time.

TESTING FOR STATIONARITY:

1. Augmented Dickey-Fuller (ADF) Test:
   - Null Hypothesis: Time series is non-stationary
   - p-value < 0.05: Reject null hypothesis (stationary)

2. KPSS Test:
   - Null Hypothesis: Time series is stationary
   - p-value > 0.05: Fail to reject null (stationary)

TRANSFORMING NON-STATIONARY TO STATIONARY:

1. Differencing:
   - Calculate differences between consecutive observations
   - Y_t' = Y_t - Y_{t-1}

2. Log Transformation:
   - Apply natural logarithm to stabilize variance
   - Y_t' = log(Y_t)

3. Seasonal Differencing:
   - For seasonal patterns: Y_t' = Y_t - Y_{t-m} (where m is seasonality period)

4. Detrending:
   - Remove trend component using moving averages or regression
"""

print(answer4)
```

```python
# Cell 6: Question 5 - Time Series Models Comparison
print("=" * 80)
print("QUESTION 5: AR, MA, ARIMA, SARIMA, SARIMAX models comparison")
print("=" * 80)

answer5 = """
TIME SERIES MODELS COMPARISON:

1. AR (AutoRegressive) Model:
   - Structure: Uses previous values to predict future values
   - Formula: Y_t = c + φ₁Y_{t-1} + φ₂Y_{t-2} + ... + ε_t
   - Application: Stock price prediction, economic forecasting

2. MA (Moving Average) Model:
   - Structure: Uses previous forecast errors to predict future values
   - Formula: Y_t = μ + ε_t + θ₁ε_{t-1} + θ₂ε_{t-2} + ...
   - Application: Quality control, signal processing

3. ARIMA (AutoRegressive Integrated Moving Average):
   - Structure: AR + I (differencing) + MA
   - Parameters: (p,d,q) where p=AR terms, d=differencing, q=MA terms
   - Application: Non-seasonal time series forecasting

4. SARIMA (Seasonal ARIMA):
   - Structure: ARIMA + seasonal components
   - Parameters: (p,d,q)(P,D,Q,s) where s=seasonal period
   - Application: Seasonal time series (monthly sales, weather data)

5. SARIMAX (Seasonal ARIMA with eXogenous variables):
   - Structure: SARIMA + external variables
   - Application: Forecasting with external factors (e.g., weather affecting sales)

KEY DIFFERENCES:
- AR/MA: Basic models for stationary data
- ARIMA: Handles non-stationary data via differencing
- SARIMA: Handles both trend and seasonality
- SARIMAX: Includes external explanatory variables
"""

print(answer5)
```

```python
# Cell 7: Question 6 - Time Series Decomposition
print("=" * 80)
print("QUESTION 6: Time Series Decomposition - AirPassengers Dataset")
print("=" * 80)

# Load and prepare data
print("Loading AirPassengers dataset...")
try:
    # Using built-in dataset
    from statsmodels.datasets import get_rdataset
    airpassengers = get_rdataset('AirPassengers')
    df = airpassengers.data
    df['date'] = pd.to_datetime(df['time'].astype(str).str.split('.').str[0] + '-' +
                               ((df['time'] - df['time'].astype(int)) * 12 + 1).astype(int).astype(str))
    df.set_index('date', inplace=True)
    ts_data = df['value']
except:
    # Fallback: Create synthetic data similar to AirPassengers
    print("Using synthetic AirPassengers-like data...")
    dates = pd.date_range('1949-01-01', '1960-12-31', freq='M')
    np.random.seed(42)
    trend = np.linspace(100, 600, len(dates))
    seasonal = 100 * np.sin(2 * np.pi * np.arange(len(dates)) / 12)
    noise = np.random.normal(0, 20, len(dates))
    ts_data = pd.Series(trend + seasonal + noise, index=dates)
    ts_data.name = 'Passengers'

print(f"Dataset shape: {ts_data.shape}")
print(f"Date range: {ts_data.index.min()} to {ts_data.index.max()}")

# Plot original series
plt.figure(figsize=(15, 10))

plt.subplot(4, 1, 1)
plt.plot(ts_data)
plt.title('Original AirPassengers Time Series')
plt.ylabel('Passengers')
plt.grid(True)

# Decompose time series
print("Decomposing time series...")
decomposition = seasonal_decompose(ts_data, model='additive', period=12)

# Plot components
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend)
plt.title('Trend Component')
plt.ylabel('Trend')
plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal)
plt.title('Seasonal Component')
plt.ylabel('Seasonality')
plt.grid(True)

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid)
plt.title('Residual Component')
plt.ylabel('Residuals')
plt.grid(True)

plt.tight_layout()
plt.show()

print("✅ Time series decomposition completed!")
```

```python
# Cell 9: Question 8 - SARIMA Forecasting
print("=" * 80)
print("QUESTION 8: SARIMA Model on AirPassengers Dataset")
print("=" * 80)

# Use the AirPassengers data from previous question
print("Preparing data for SARIMA modeling...")

# Split data into train and test
train_size = int(len(ts_data) * 0.8)
train_data = ts_data[:train_size]
test_data = ts_data[train_size:]

print(f"Training data: {len(train_data)} months")
print(f"Test data: {len(test_data)} months")

# Fit SARIMA model
print("Fitting SARIMA model...")
try:
    # Using simpler parameters for demonstration
    sarima_model = SARIMAX(train_data,
                          order=(1, 1, 1),           # (p,d,q) - non-seasonal
                          seasonal_order=(1, 1, 1, 12), # (P,D,Q,s) - seasonal
                          enforce_stationarity=False,
                          enforce_invertibility=False)

    sarima_result = sarima_model.fit(disp=False)
    print("✅ SARIMA model fitted successfully!")

    # Forecast next 12 months
    forecast_steps = 12
    forecast = sarima_result.get_forecast(steps=forecast_steps)
    forecast_mean = forecast.predicted_mean
    confidence_int = forecast.conf_int()

    # Plot results
    plt.figure(figsize=(15, 8))

    # Plot historical data
    plt.plot(train_data.index, train_data, label='Training Data', color='blue')
    plt.plot(test_data.index, test_data, label='Actual Test Data', color='green')

    # Plot forecast
    forecast_index = pd.date_range(start=test_data.index[0], periods=forecast_steps, freq='M')
    plt.plot(forecast_index, forecast_mean, label='SARIMA Forecast', color='red', linestyle='--')

    # Plot confidence interval
    plt.fill_between(forecast_index,
                    confidence_int.iloc[:, 0],
                    confidence_int.iloc[:, 1],
                    color='pink', alpha=0.3, label='95% Confidence Interval')

    plt.title('SARIMA Model: 12-Month Forecast for AirPassengers')
    plt.xlabel('Date')
    plt.ylabel('Passengers')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    # Print forecast values
    print("\n📊 Forecasted values for next 12 months:")
    for i, (date, value) in enumerate(zip(forecast_index, forecast_mean), 1):
        print(f"Month {i:2d}: {date.strftime('%Y-%m')} - {value:.1f} passengers")

except Exception as e:
    print(f"❌ Error in SARIMA modeling: {e}")
    print("This might be due to complex seasonal patterns. Try different parameters.")
```

```python
# Cell 10: Question 9 - Local Outlier Factor (LOF)
print("=" * 80)
print("QUESTION 9: Local Outlier Factor Anomaly Detection")
print("=" * 80)

# Create a synthetic dataset with clusters and outliers
print("Creating synthetic dataset for LOF...")
np.random.seed(42)

# Generate clustered data
X1, y1 = make_blobs(n_samples=300, centers=1, cluster_std=0.5, center_box=(-5, 5))
X2, y2 = make_blobs(n_samples=300, centers=1, cluster_std=0.8, center_box=(5, 5))

# Combine clusters
X = np.vstack([X1, X2])

# Add some outliers
outliers = np.random.uniform(low=-10, high=10, size=(20, 2))
X = np.vstack([X, outliers])

print(f"Dataset shape: {X.shape}")

# Apply Local Outlier Factor
print("Applying Local Outlier Factor...")
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# Calculate LOF scores
lof_scores = lof.negative_outlier_factor_

# Create results dataframe
lof_data = pd.DataFrame({
    'x1': X[:, 0],
    'x2': X[:, 1],
    'anomaly': y_pred,
    'lof_score': lof_scores
})

# Convert anomaly labels (1=normal, -1=anomaly)
lof_data['anomaly'] = lof_data['anomaly'].map({1: 0, -1: 1})

# Separate normal and anomaly points
normal_points = lof_data[lof_data['anomaly'] == 0]
anomaly_points = lof_data[lof_data['anomaly'] == 1]

# Visualize results
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.scatter(normal_points['x1'], normal_points['x2'],
           c=normal_points['lof_score'], cmap='viridis',
           alpha=0.6, s=50, label='Normal')
plt.scatter(anomaly_points['x1'], anomaly_points['x2'],
           c='red', alpha=0.8, s=80, label='Anomaly', edgecolors='black')
plt.colorbar(label='LOF Score (negative)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Local Outlier Factor - Anomaly Detection')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
# Plot LOF scores distribution
plt.hist(lof_scores, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(x=lof_scores[anomaly_points.index].mean(), color='red', linestyle='--',
           label=f'Anomaly Threshold')
plt.xlabel('LOF Score (negative)')
plt.ylabel('Frequency')
plt.title('Distribution of LOF Scores')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Detected {anomaly_points.shape[0]} anomalies using LOF")
print("✅ Local Outlier Factor anomaly detection completed!")
```

```python
# Cell 1: Corrected Power Grid Monitoring Solution
print("=" * 80)
print("QUESTION 10: Power Grid Monitoring - Real-time Solution")
print("=" * 80)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("🔄 Creating realistic power grid dataset...")

# Generate 7 days of data at 15-minute intervals
dates = pd.date_range('2024-01-01', '2024-01-07', freq='15T')
n_points = len(dates)

# Base energy demand pattern (daily seasonality)
base_demand = 800 + 300 * np.sin(2 * np.pi * np.arange(n_points) / 96)  # 96 intervals per day

# Add weekly pattern (lower on weekends)
day_of_week = pd.Series(dates).dt.dayofweek
weekly_pattern = np.where(day_of_week >= 5, -100, 0)  # Weekend reduction

# Weather factors
temperature = 20 + 8 * np.sin(2 * np.pi * np.arange(n_points) / 96) + np.random.normal(0, 2, n_points)
humidity = 50 + 15 * np.sin(2 * np.pi * np.arange(n_points) / 96) + np.random.normal(0, 3, n_points)

# Create final energy demand with weather influence
energy_demand = (base_demand + weekly_pattern +
                10 * temperature +  # Higher temp = more AC usage
                5 * humidity +      # Humidity affects systems
                np.random.normal(0, 30, n_points))

# Add realistic anomalies - CORRECTED THIS LINE
anomaly_indices = [50, 150, 300, 400, 450, 500, 550] # Adjusted indices to be within bounds

# Spikes (equipment failure, sudden demand surge)
energy_demand[anomaly_indices[:3]] = [energy_demand[i] * 2.0 for i in anomaly_indices[:3]]
# Drops (power outage, equipment shutdown) - CORRECTED: [3:5] not [:3:5]
energy_demand[anomaly_indices[3:5]] = [energy_demand[i] * 0.3 for i in anomaly_indices[3:5]]
# Surges (special events, industrial activity)
energy_demand[anomaly_indices[5:]] = [energy_demand[i] + 500 for i in anomaly_indices[5:]]

# Create comprehensive dataset
power_data = pd.DataFrame({
    'timestamp': dates,
    'energy_demand': energy_demand,
    'temperature': temperature,
    'humidity': humidity,
    'hour': pd.Series(dates).dt.hour,
    'day_of_week': day_of_week,
    'is_weekend': (day_of_week >= 5).astype(int)
})

# Mark actual anomalies for evaluation
power_data['true_anomaly'] = 0
power_data.loc[anomaly_indices, 'true_anomaly'] = 1

power_data.set_index('timestamp', inplace=True)

print(f"✅ Dataset created: {power_data.shape} records")
print(f"📅 Date range: {power_data.index.min()} to {power_data.index.max()}")
print(f"🔥 Anomalies injected: {len(anomaly_indices)}")
print(f"📊 Data sample:")
print(power_data.head())
```

```python
# Cell 2: Real-time Anomaly Detection - CORRECTED
print("\n" + "="*50)
print("REAL-TIME ANOMALY DETECTION IMPLEMENTATION")
print("="*50)

# Feature engineering for anomaly detection
features_for_anomaly = ['energy_demand', 'temperature', 'humidity', 'hour', 'is_weekend']
X_anomaly = power_data[features_for_anomaly].copy()

# Add rolling statistics
X_anomaly['demand_rolling_mean_2h'] = power_data['energy_demand'].rolling(window=8).mean()
X_anomaly['demand_rolling_std_2h'] = power_data['energy_demand'].rolling(window=8).std()
X_anomaly = X_anomaly.fillna(method='bfill')

print("📊 Features for anomaly detection:")
print(X_anomaly.columns.tolist())

# Check for NaN values
print(f"NaN values in features: {X_anomaly.isna().sum().sum()}")

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_anomaly)

# Apply Isolation Forest
print("\n🔍 Applying Isolation Forest...")
iso_forest = IsolationForest(
    n_estimators=100,  # Reduced for faster execution
    contamination=0.03,
    random_state=42,
    max_samples=256
)

anomaly_predictions = iso_forest.fit_predict(X_scaled)
power_data['predicted_anomaly'] = (anomaly_predictions == -1).astype(int)

# Calculate performance
true_anomalies = power_data['true_anomaly']
predicted_anomalies = power_data['predicted_anomaly']

tp = ((true_anomalies == 1) & (predicted_anomalies == 1)).sum()
fp = ((true_anomalies == 0) & (predicted_anomalies == 1)).sum()
fn = ((true_anomalies == 1) & (predicted_anomalies == 0)).sum()

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"📈 Anomaly Detection Results:")
print(f"   - Precision: {precision:.3f}")
print(f"   - Recall: {recall:.3f}")
print(f"   - F1-Score: {f1_score:.3f}")
print(f"   - True Anomalies Detected: {tp}/{len(anomaly_indices)}")
print(f"   - False Positives: {fp}")
```

```python
# Cell 3: Simple Forecasting Implementation
print("\n" + "="*50)
print("TIME SERIES FORECASTING IMPLEMENTATION")
print("="*50)

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# Simple moving average forecasting
def simple_forecast(data, window_size=96):  # 24 hours window
    forecasts = []
    for i in range(len(data) - window_size):
        window = data.iloc[i:i+window_size]
        # Use mean of last window as forecast
        forecast = window.mean()
        forecasts.append(forecast)
    return forecasts

# Prepare data
train_size = int(len(power_data) * 0.8)
train_data = power_data.iloc[:train_size]
test_data = power_data.iloc[train_size:]

print(f"📊 Data Split:")
print(f"   - Training: {len(train_data)} records")
print(f"   - Testing: {len(test_data)} records")

# Generate forecasts
forecast_window = 96  # 24 hours
forecasts = simple_forecast(train_data['energy_demand'], forecast_window)

# Align forecasts with test data (take only the needed portion)
if len(forecasts) > len(test_data):
    forecasts = forecasts[:len(test_data)]

# Create forecast series
forecast_index = test_data.index[:len(forecasts)]
forecast_series = pd.Series(forecasts, index=forecast_index)

# Calculate accuracy
actual_values = test_data['energy_demand'][:len(forecasts)]
mae = mean_absolute_error(actual_values, forecast_series)
mape = mean_absolute_percentage_error(actual_values, forecast_series)

print(f"📈 Forecasting Performance:")
print(f"   - MAE: {mae:.2f} MW")
print(f"   - MAPE: {mape:.3f}")

# Add confidence intervals
confidence_upper = forecast_series * 1.15
confidence_lower = forecast_series * 0.85
```
